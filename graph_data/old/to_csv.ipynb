```python
import pandas as pd
```

```python
file = open('hubs.txt')
data = file.read()
data
```

```python
data = data.split('\n')
data
```

```python
data = list(map(lambda x: x.split(), data))
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df = df.rename(columns={0: "type", 1: "x", 2: 'y'})
df
```

```python
df = df.drop(3, axis=1)
df
```

```python
df.to_csv('../nodes.csv', index=False)
```
